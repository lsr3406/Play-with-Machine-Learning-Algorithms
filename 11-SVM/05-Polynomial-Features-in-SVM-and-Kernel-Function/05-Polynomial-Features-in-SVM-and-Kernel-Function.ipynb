# SVM中的使用多项式特征

```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
X, y = datasets.make_moons(noise=0.1)
axis = [
    X[:, 0].min() * 1.1 - X[:, 0].max() * 0.1,
    X[:, 0].max() * 1.1 - X[:, 0].min() * 0.1,
    X[:, 1].min() * 1.1 - X[:, 1].max() * 0.1,
    X[:, 1].max() * 1.1 - X[:, 1].min() * 0.1,
]
print(X.shape)
print(y.shape)
```

```python
plt.scatter(X[y==0, 0], X[y==0, 1], color="#f44336")
plt.scatter(X[y==1, 0], X[y==1, 1], color="#ff9800")
plt.show()
```

#### 使用多项式特征的SVM

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline

def PolynomialSVC(degree, C):
    return Pipeline([
        ("poly", PolynomialFeatures(degree=degree)),
        ("std_scaler", StandardScaler()),
        ("svm_clf", LinearSVC(C=C))
    ])
```

```python
poly_svc = PolynomialSVC(degree=3, C=1e8)
poly_svc.fit(X, y)
poly_svc.score(X, y)
```

```python
def plot_svc_decision_boundary(model, axis):
    cx = np.linspace(axis[0], axis[1], 100*(axis[1]-axis[0]))
    cy = np.linspace(axis[2], axis[3], 100*(axis[3]-axis[2]))
    CX, CY = np.meshgrid(cx, cy)

    new_X = np.c_[CX.ravel(), CY.ravel()]
    CZ = model.predict(new_X).reshape(CX.shape)

    from matplotlib.colors import ListedColormap
    custom_cmap = ListedColormap(["#f9a19a", "#ffcb7f"])
    plt.contourf(CX, CY, CZ, linewidth=5, cmap=custom_cmap)
```

```python
plot_svc_decision_boundary(poly_svc, axis)
plt.scatter(X[y==0, 0], X[y==0, 1], color="#f44336")
plt.scatter(X[y==1, 0], X[y==1, 1], color="#ff9800")
plt.show()
```

```python
poly_svc = PolynomialSVC(degree=3, C=1)
poly_svc.fit(X, y)
poly_svc.score(X, y)

plot_svc_decision_boundary(poly_svc, axis)
plt.scatter(X[y==0, 0], X[y==0, 1], color="#f44336")
plt.scatter(X[y==1, 0], X[y==1, 1], color="#ff9800")
plt.show()
```

```python
poly_svc = PolynomialSVC(degree=3, C=0.1)
poly_svc.fit(X, y)
poly_svc.score(X, y)

plot_svc_decision_boundary(poly_svc, axis)
plt.scatter(X[y==0, 0], X[y==0, 1], color="#f44336")
plt.scatter(X[y==1, 0], X[y==1, 1], color="#ff9800")
plt.show()
```

#### 使用多项式核函数的SVM

```python
from sklearn.svm import SVC

def PolynomialKernelSVC(degree, C):
    return Pipeline([
        ("std_scaler", StandardScaler()),
        ("svc_clf", SVC(kernel="poly", degree=degree, C=C))
    ])
```

```python
poly_kernel_svc = PolynomialKernelSVC(3, 1e3)
poly_kernel_svc.fit(X, y)
poly_kernel_svc.score(X, y)

plot_svc_decision_boundary(poly_svc, axis)
plt.scatter(X[y==0, 0], X[y==0, 1], color="#f44336")
plt.scatter(X[y==1, 0], X[y==1, 1], color="#ff9800")
plt.show()
```
