# CART 和 决策树的超参数

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets as datasets
```

```python
X, y = datasets.make_moons(noise=0.25, random_state=666)

plt.scatter(X[y==0, 0], X[y==0, 1], color="#f44336")
plt.scatter(X[y==1, 0], X[y==1, 1], color="#ff9800")
plt.show()

axis = [
    X[:, 0].min() * 1.1 - X[:, 0].max() * 0.1,
    X[:, 0].max() * 1.1 - X[:, 0].min() * 0.1,
    X[:, 1].min() * 1.1 - X[:, 1].max() * 0.1,
    X[:, 1].max() * 1.1 - X[:, 1].min() * 0.1,
]
```

```python
def plot_decision_boundary(algo, axis):
    
    x0 = np.linspace(axis[0], axis[1], int(100 * (axis[1]-axis[0])))
    x1 = np.linspace(axis[2], axis[3], int(100 * (axis[3]-axis[2])))
    x0, x1 = np.meshgrid(x0, x1)
    
    X_new = np.c_[x0.ravel(), x1.ravel()]
    y_predict = algo.predict(X_new)
    z = y_predict.reshape(x0.shape)
    
    from matplotlib.colors import ListedColormap
    custom_map = ListedColormap(["#f9a19a", "#ffcb7f"])
    
    plt.contourf(x0, x1, z, linewidth=5, cmap=custom_map)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_clf = DecisionTreeClassifier()
dt_clf.fit(X, y)
```

```python
plot_decision_boundary(dt_clf, axis)
plt.scatter(X[y==0, 0], X[y==0, 1], color="#f44336")
plt.scatter(X[y==1, 0], X[y==1, 1], color="#ff9800")
plt.show()
```

```python
dt_clf = DecisionTreeClassifier(max_depth=2)
dt_clf.fit(X, y)

plot_decision_boundary(dt_clf, axis)
plt.scatter(X[y==0, 0], X[y==0, 1], color="#f44336")
plt.scatter(X[y==1, 0], X[y==1, 1], color="#ff9800")
plt.show()
```

```python
dt_clf = DecisionTreeClassifier(min_samples_split=10)
dt_clf.fit(X, y)

plot_decision_boundary(dt_clf, axis)
plt.scatter(X[y==0, 0], X[y==0, 1], color="#f44336")
plt.scatter(X[y==1, 0], X[y==1, 1], color="#ff9800")
plt.show()
```

```python
dt_clf = DecisionTreeClassifier(min_samples_leaf=6)
dt_clf.fit(X, y)

plot_decision_boundary(dt_clf, axis)
plt.scatter(X[y==0, 0], X[y==0, 1], color="#f44336")
plt.scatter(X[y==1, 0], X[y==1, 1], color="#ff9800")
plt.show()
```

```python
dt_clf = DecisionTreeClassifier(max_leaf_nodes=4)
dt_clf.fit(X, y)

plot_decision_boundary(dt_clf, axis)
plt.scatter(X[y==0, 0], X[y==0, 1], color="#f44336")
plt.scatter(X[y==1, 0], X[y==1, 1], color="#ff9800")
plt.show()
```
