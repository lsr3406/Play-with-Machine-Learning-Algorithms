# 模型复杂度曲线

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
X, y = boston.data, boston.target

X = X[y < 50]
y = y[y < 50]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

#### min_samples_leaf

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
```

```python
maxSampleLeaf = 506
train_scores = []
test_scores = []

for k in range(1, maxSampleLeaf+1):
    dt_reg = DecisionTreeRegressor(min_samples_leaf=k)
    dt_reg.fit(X_train, y_train)
    y_train_predict = dt_reg.predict(X_train)
    train_scores.append(r2_score(y_train, y_train_predict))
    test_scores.append(dt_reg.score(X_test, y_test))
train_scores = np.array(train_scores)
test_scores = np.array(test_scores)

plt.plot(np.arange(1, maxSampleLeaf + 1), train_scores, label="train", color="#f44336")
plt.plot(np.arange(1, maxSampleLeaf + 1), test_scores, label="test", color="#ff9800")
plt.legend()
plt.axis([-0.1*maxSampleLeaf, 1.1*maxSampleLeaf, -0.1, 1.1])
plt.show()
```
