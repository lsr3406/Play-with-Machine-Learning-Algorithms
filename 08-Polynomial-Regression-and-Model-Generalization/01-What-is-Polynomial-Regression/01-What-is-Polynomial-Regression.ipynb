# 什么是多项式回归

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(-3, 3, size=100)
X = x.reshape(-1, 1)
y = 0.5 * x ** 2 + x - 2 + np.random.normal(0, 1, 100)
```

```python
plt.scatter(x, y, marker='.')
plt.show()
```

#### 线性回归

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X, y)
# k = reg.coef_
# b = reg.intercept_
```

```python
cx = np.linspace(-3, 3, 121)
# cy = k * cx + b
cy = reg.predict(cx.reshape(-1, 1))

plt.scatter(x, y, marker='.')
plt.plot(cx, cy)
plt.show()
```

#### 解决方案， 添加一个特征

```python
X2 = np.hstack([X, X ** 2])

reg = LinearRegression()
reg.fit(X2, y)
b, a = reg.coef_
c = reg.intercept_
print(a, b, c)
```

```python
cx = np.linspace(-3, 3, 121)
cy = a * cx ** 2 + b * cx + c

plt.scatter(x, y, marker='.')
plt.plot(cx, cy)
plt.show()
```
