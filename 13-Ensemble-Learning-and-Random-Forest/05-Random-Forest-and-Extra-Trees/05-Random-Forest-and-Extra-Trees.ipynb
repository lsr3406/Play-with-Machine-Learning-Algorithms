# 随机森林

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=500, noise=0.25)
```

```python
plt.scatter(X[y==0,0], X[y==0,1], color="#f44336")
plt.scatter(X[y==1,0], X[y==1,1], color="#ff9800")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(n_estimators=500,
                                oob_score=True)
rf_clf.fit(X, y)
rf_clf.oob_score_
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(n_estimators=500,
                                oob_score=True,
                                max_leaf_nodes=16)
rf_clf.fit(X, y)
rf_clf.oob_score_
```

随机森林拥有决策树和BaggingClassifier的所有参数

#### Extra-Trees

```python
from sklearn.ensemble import ExtraTreesClassifier
et_clf = ExtraTreesClassifier(n_estimators=500,
                             bootstrap=True,
                             oob_score=True)
et_clf.fit(X, y)
et_clf.oob_score_
```
