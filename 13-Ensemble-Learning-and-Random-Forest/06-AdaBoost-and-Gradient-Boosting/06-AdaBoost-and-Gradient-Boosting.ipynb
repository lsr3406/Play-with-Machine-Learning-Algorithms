# Boosting

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_moons
X, y = make_moons(noise=0.25, n_samples=500)
```

```python
plt.scatter(X[y==0,0], X[y==0,1], color="#f44336")
plt.scatter(X[y==1,0], X[y==1,1], color="#ff9800")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

#### AdaBoosting

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
ada_clf = AdaBoostClassifier(DecisionTreeClassifier(),
                            n_estimators=500)
ada_clf.fit(X_train, y_train)
ada_clf.score(X_test, y_test)
```

#### Gradient Boosting

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
gb_clf = GradientBoostingClassifier(max_depth=2,
                            n_estimators=30)
gb_clf.fit(X_train, y_train)
gb_clf.score(X_test, y_test)
```
