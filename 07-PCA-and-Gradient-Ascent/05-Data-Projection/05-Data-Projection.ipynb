# 从高维数据向低维数据的映射

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.empty((1000, 2))
X[:, 0] = np.random.uniform(-1, 1, (1000,))
X[:, 1] = 0.75 * X[:, 0] + np.sin(12.56 * X[:, 0]) * 0.2
X += np.random.normal(0, 0.02, (1000,2))

plt.scatter(X[:,0], X[:, 1], marker=".")
plt.show()
```

```python
from playML.PCA import PCA
```

```python
pca = PCA(n_components=2)
pca.fit(X)
```

```python
pca.components_
```

```python
X_reduced = pca.transform(X)
plt.scatter(X_reduced[:,0], X_reduced[:, 1], marker=".")
plt.show()
```

```python
X_restore = pca.inverse_transform(X_reduced)
plt.scatter(X_restore[:,0], X_restore[:, 1], marker=".")
plt.show()
```

#### 降噪实验

```python
pca = PCA(n_components=1)
pca.fit(X)
X_reduced = pca.transform(X)
X_restore = pca.inverse_transform(X_reduced)
```

```python
plt.scatter(X[:, 0], X[:, 1], marker=".")
plt.scatter(X_restore[:, 0], X_restore[:, 1], marker=".")
plt.show()
```

#### 使用 sklearn 中的 PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(X[:100])
X_reduction = pca.transform(X[:100])
X_restore = pca.inverse_transform(X_reduction)
```

```python
plt.scatter(X[:100, 0], X[:100, 1], marker="o")
plt.scatter(X_reduction[:, 0], X_reduction[:, 1], marker="+")
plt.scatter(X_restore[:, 0], X_restore[:, 1], marker=".")
plt.show()
```
