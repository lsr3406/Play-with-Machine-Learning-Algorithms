# 模拟梯度下降法

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 6, 141)
y = (x - 2.5) ** 2 - 1

plt.plot(x, y, color="#f44336")
plt.show()
```

```python
def J(x):
    return (x - 2.5) ** 2 - 1
def dJ(x):
    return 2 * (x - 2.5)
```

```python
epsilon = 1e-8
eta = 0.1

theta = 0.0
while True:
    gradient = dJ(theta)
    last_theta = theta
    theta -= eta * gradient
    
    if abs(J(last_theta) - J(theta)) <= epsilon:
        break

print(theta)
print(J(theta))
```

```python
epsilon = 1e-8
eta = 0.1
theta = 0.0
n_iter = 0
theta_list = [theta]

last_J = J(theta)
while True:
    gradient = dJ(theta)
    theta -= eta * gradient
    theta_list.append(theta)
    
    current_J = J(theta)
    if abs(last_J - current_J) <= epsilon or n_iter >= 1e2:
        break
    last_J = current_J
    n_iter += 1
    
print(theta)
print(current_J)

theta_list = np.array(theta_list)

plt.plot(x, y, color="#f44336")
plt.plot(theta_list, J(theta_list), color="#e91e63", marker="+")
plt.show()
```
